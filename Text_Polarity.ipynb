```python
import pandas as pd
from textblob import TextBlob
```

```python
text_df = pd.read_csv("polarity_check.csv")
```

```python
text_df.head()
```

```python
text_df.Text
```

```python
texts = text_df.Text
```

```python
# Check polarity
data = []
for text in texts:
    polarity = TextBlob(text).sentiment.polarity
    sentiment = "Positive" if polarity > 0 else "Negative" if polarity < 0 else "Neutral"
    print(f"Text: {text}\nPolarity: {polarity} ({sentiment})\n")
    data.append([text, polarity, sentiment])
```

```python
data
```

```python
df = pd.DataFrame(data, columns=["Text", "Polarity", "Sentiment"])
```

```python
df.reset_index(inplace = True)
```

```python
df.head(), text_df.head()
```

```python
final_text = pd.merge(text_df, df, left_on="ID", right_on="index", how="left")  # Inner join
```

```python
final_text.to_csv('final.csv')
```
